```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
```

```python
df = pd.read_excel('a.xlsx')
```

```python
df = df.rename(columns = {'Impress√µes dos banners' : 'Imp'})
```

```python
df = df.drop(df.columns[[5, 6, 7]], axis = 1)
```

```python
df.head()
```

```python
dfA = df.query('Anunciante == "A"')
dfB = df.query('Anunciante == "B"')
dfC = df.query('Anunciante == "C"')
dfD = df.query('Anunciante == "D"')
```

```python
plt.style.use('ggplot')
```

```python
plt.figure(figsize = (10, 6))
plt.plot(dfA['Data'], dfA['CPM'], c = 'cyan', label = 'A')
plt.plot(dfB['Data'], dfB['CPM'], c = 'magenta', label = 'B')
plt.plot(dfC['Data'], dfC['CPM'], c = 'black', label = 'C')
plt.plot(dfD['Data'], dfD['CPM'], c = 'salmon', label = 'D')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10, 6))
plt.plot(dfA['Data'], dfA['Imp'], c = 'cyan', label = 'A')
plt.plot(dfB['Data'], dfB['Imp'], c = 'magenta', label = 'B')
plt.plot(dfC['Data'], dfC['Imp'], c = 'black', label = 'C')
plt.plot(dfD['Data'], dfD['Imp'], c = 'salmon', label = 'D')
plt.legend()
plt.show()
```
